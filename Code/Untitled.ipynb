```python
f1 = t_ratio_parameter_estimation.log_f_a_min
f2 = t_ratio_parameter_estimation.log_f_a_max
mu_1 = t_ratio_parameter_estimation.log_mu_min
mu_2 = t_ratio_parameter_estimation.log_mu_max

def intersect_interval(i1, i2):
    (a1, a2) = i1
    (b1, b2) = i2
    return np.where(((a1 < b1) & (a2 < b1)) | ((a1 > b2) & (a2 > b2)),
        0,
        np.minimum(a2, b2) - np.maximum(a1, b1)
    )

def P(log_m_a):
    return (
        intersect_interval(
            (mu_1, mu_2),
            ((f1 + np.log10(2) + log_m_a) / 4, (f2 + np.log10(2) + log_m_a) / 4)
        ) /
        ((f2 - f1) * (mu_2 - mu_1)) 
    )

log_m_a = np.linspace(-10, -25, 300)

plt.plot(log_m_a, P(log_m_a))
```
