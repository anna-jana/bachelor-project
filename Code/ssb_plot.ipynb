```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

N = 50
R = 2.1
x = np.linspace(-R, R, N)
y = np.linspace(-R, R, N)
X, Y = np.meshgrid(x, y)
r = np.sqrt(X**2 + Y**2)
Z = -10 * r**2 + r**4

# Plot a basic wireframe.
ax.plot_surface(X, Y, Z, alpha=0.5) # , rstride=10, cstride=10)
ax.axis("off")
```

```python
N = 50
r = np.linspace(0, 3, 30)
phi = np.linspace(0, 2*np.pi, 30)
R, PHI = np.meshgrid(r, phi)
X = R * np.cos(PHI)
Y = R * np.sin(PHI)
alpha = 10
Z = -alpha * R**2 + R**4

fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 3, 1, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.2)
ax.plot3D([0], [0], [0], color="red", marker="o", markersize=10, alpha=1)
ax.axis("off")

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.2)
ax.plot3D([np.sqrt(alpha/2)], [0], [- 1 / 4 * alpha**2], color="red", marker="o", markersize=10, alpha=1)
ax.axis("off")

ax = fig.add_subplot(1, 3, 3, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.2)
ax.plot3D([np.sqrt(alpha/2)], [0], [- 1 / 4 * alpha**2], color="red", marker="o", markersize=10, alpha=1)
ax.axis("off")
ax.view_init(15, 0)
```

```python
N = 50
r = np.linspace(0, 3, 30)
phi = np.linspace(0, 2*np.pi, 30)
R, PHI = np.meshgrid(r, phi)
X = R * np.cos(PHI)
Y = R * np.sin(PHI)
alpha = 10
Z = -alpha * R**2 + R**4

x_broken = np.sqrt(alpha/2)
y_broken = 0
z_broken = - 1 / 4 * alpha**2

fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 3, 1, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.2)
ax.plot3D([0], [0], [0], color="red", marker="o", markersize=10, alpha=1)
ax.axis("off")
ax.view_init(0, 0)
ax.set_title("a) Unbroken Symmetry", fontsize=20)

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.2)
ax.plot3D([0], [0], [0], color="red", marker="$o$", markersize=10)
ax.plot3D([x_broken], [y_broken], [z_broken], color="red", marker="o", markersize=10, alpha=1)
# ax.quiver3D([0], [0], [0], [x_broken], [y_broken], [z_broken], arrow_length_ratio=0.1)
ax.axis("off")
ax.view_init(0, 90)
ax.set_title("b) Spontaneous Symmetry Breaking", fontsize=20)

ax = fig.add_subplot(1, 3, 3, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.2)
ax.plot3D([x_broken], [y_broken], [z_broken], color="red", marker="o", markersize=10, alpha=1)
ax.quiver3D([x_broken], [y_broken], [z_broken], [-1], [-1], [0], linewidths=[1], arrow_length_ratio=0.5, color="red")
ax.axis("off")
ax.view_init(-40, 90)
ax.set_title("c) Axion acquires Mass via Instanton Effects", fontsize=20)

plt.tight_layout()
import config
plt.savefig(config.plot_path + "/ssb.pdf")
```
