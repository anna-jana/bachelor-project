```python
import numpy as np
import matplotlib.pyplot as plt
import config
```

```python
plt.figure(figsize=(8, 4))
fig = plt.gcf()
ax = plt.gca()
theta = np.linspace(-np.pi, np.pi, 400)
lw = 2.4
plt.plot(theta, 0.5*theta**2, label=r"$\sim \frac{1}{2} \theta^2$", linestyle="-", linewidth=lw)
plt.plot(theta, 1 - np.cos(theta), label=r"$\sim 1 - \cos(\theta)$", linestyle="--", linewidth=lw)
plt.plot(theta, 4 * (1 - np.sqrt(2 + 2 * np.cos(theta)) / 2), 
         label=r"$\sim 1 - \sqrt{\frac{1 + \cos{\theta}}{2}}$", 
         linestyle="-.", linewidth=lw)
plt.ylim(-1, 3)
plt.xlabel(r"$\theta$", fontsize=15)
plt.ylabel(r"$V(\theta) \, / \, \mathrm{a.u.}$", fontsize=15)
plt.xticks(np.arange(-np.pi, np.pi + np.pi/4, np.pi/4),
           [r"$-\pi$",r"$-\frac{3}{4}\pi$",r"$-\frac{1}{2}\pi$",r"$-\frac{1}{4}\pi$",
            "0"] +
            list(reversed([r"$\pi$",r"$\frac{3}{4}\pi$",r"$\frac{1}{2}\pi$",r"$\frac{1}{4}\pi$"])))
ax.tick_params(labelsize=15)
plt.yticks([0], ["0"])
plt.legend(loc="lower center", fancybox=False, ncol=3, shadow=False, borderaxespad=1, frameon=False, fontsize=15) 
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
plt.tight_layout()
plt.savefig(config.plot_path + "/potential_plot.pdf")
```
