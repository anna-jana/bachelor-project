```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1)
phi = np.linspace(-np.pi, np.pi, 400)
plt.plot(phi, 0.5*phi**2, label=r"$\sim \frac{1}{2} \phi^2$", linestyle="--")
plt.plot(phi, 1 - np.cos(phi), label=r"$\sim 1 - \cos(\phi)$")
plt.ylim(-1, 3)
plt.xlabel(r"$\phi$", fontsize=15)
plt.ylabel(r"$\sim V(\phi)$", fontsize=15)
plt.xticks(np.arange(-np.pi, np.pi + np.pi/4, np.pi/4),
           [r"$-\pi$",r"$-\frac{3}{4}\pi$",r"$-\frac{1}{2}\pi$",r"$-\frac{1}{4}\pi$",
            "0"] +
            list(reversed([r"$\pi$",r"$\frac{3}{4}\pi$",r"$\frac{1}{2}\pi$",r"$\frac{1}{4}\pi$"])))
plt.yticks([0], ["0"])
plt.legend(loc="lower center", fancybox=False, ncol=2, shadow=False, borderaxespad=1, frameon=True) 
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
plt.tight_layout()
plt.savefig("Plots/potential_plot.pdf")
```
